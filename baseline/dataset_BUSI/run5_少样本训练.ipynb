```python
import os
import random

# 1. 设置路径
dataset_root = "/home/zhengsongming/jupyterworkspace/datasets/BUSI_for_SAM2/ImageSets"
original_train_file = os.path.join(dataset_root, "train.txt")

# 2. 读取原始列表
with open(original_train_file, 'r') as f:
    lines = f.readlines()

total_samples = len(lines)
random.seed(42) # 固定种子，保证可复现

# 3. 生成不同比例的子集
ratios = [0.1, 0.25, 0.5]

for ratio in ratios:
    # 计算数量
    num_samples = int(total_samples * ratio)
    # 随机抽样
    subset = random.sample(lines, num_samples)
    
    # 保存文件名，例如 train_0.1.txt
    new_filename = f"train_{int(ratio*100)}.txt"
    save_path = os.path.join(dataset_root, new_filename)
    
    with open(save_path, 'w') as f:
        f.writelines(subset)
    
    print(f"Generated {new_filename}: {len(subset)} samples")
```
