```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 准备数据
ratios = [10, 25, 50, 100]  # X轴数值
labels = ["10%", "25%", "50%", "100%"]
dice_ours = [91.91, 92.66, 93.19, 93.46]
dice_medsam2 = [91.20, 92.61, 92.90, 93.27]
dice_sam2 = [87.23, 87.63, 89.73, 89.93]
dice_imedsam = [86.32, 87.35, 88.67, 88.56]
dice_unet = [43.97, 60.35, 65.67, 77.85]
dice_swin_unet = [43.86, 58.56, 67.17, 81.15]

# 2. 设置绘图风格 (学术论文风格)
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['xtick.major.width'] = 1.2
plt.rcParams['ytick.major.width'] = 1.2

fig, ax = plt.subplots(figsize=(7, 5.5), dpi=150)

# 3. 绘制曲线
# 统一设置：所有点都用 'x'
marker_style = 'x' 
linewidth = 1.8
base_markersize = 8    # 基础标记大小
base_mew = 2.0         # markeredgewidth (标记线条宽度，因为x比较细，需要加粗)

# --- Ours (最突出: 红色, 更大的x, 更粗的线) ---
ax.plot(ratios, dice_ours, marker=marker_style, color='#D62728', label='Ours', 
        linewidth=2.5, markersize=10, markeredgewidth=3.0, zorder=10) 

# --- SAM variants (同类竞品: 实线) ---
ax.plot(ratios, dice_medsam2, marker=marker_style, color='#1F77B4', label='MedSAM2', 
        linewidth=linewidth, markersize=base_markersize, markeredgewidth=base_mew)

ax.plot(ratios, dice_sam2, marker=marker_style, color='#2CA02C', label='SAM2', 
        linewidth=linewidth, markersize=base_markersize, markeredgewidth=base_mew)

ax.plot(ratios, dice_imedsam, marker=marker_style, color='#9467BD', label='I-MedSAM', 
        linewidth=linewidth, markersize=base_markersize, markeredgewidth=base_mew)

# --- Traditional Methods (Baseline: 虚线) ---
ax.plot(ratios, dice_swin_unet, marker=marker_style, color='#FF7F0E', linestyle='--', label='Swin-Unet', 
        linewidth=linewidth, markersize=base_markersize, markeredgewidth=base_mew)

ax.plot(ratios, dice_unet, marker=marker_style, color='#7F7F7F', linestyle='--', label='U-Net', 
        linewidth=linewidth, markersize=base_markersize, markeredgewidth=base_mew)

# 4. 美化图表
# 设置X轴
ax.set_xticks(ratios)
ax.set_xticklabels(labels, fontsize=12)
ax.set_xlabel('Proportion of Training Annotation Amount (%)', fontsize=14, labelpad=10)

# 设置Y轴
ax.set_ylabel('Dice (%)', fontsize=14, labelpad=10)
ax.set_ylim(40, 98) 
ax.tick_params(axis='y', labelsize=12)

# 网格线
ax.grid(True, which='major', linestyle='-', alpha=0.6, color='lightgray')

# 图例 (Legend) - 放大并调整位置
# fontsize=13 (放大), borderpad=0.6 (增加内边距让框看起来大一点)
ax.legend(loc='lower right', fontsize=13, frameon=True, 
          edgecolor='black', framealpha=0.9, fancybox=False, borderpad=0.6)

# 5. 显示和保存
plt.tight_layout()

# 保存
# plt.savefig('data_efficiency_plot_v2.pdf', format='pdf', bbox_inches='tight')
plt.savefig('data_efficiency_plot_v2.png', format='png', dpi=500, bbox_inches='tight')

plt.show()
```
